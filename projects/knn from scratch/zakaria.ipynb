Generating 4 datasets with these functions and ploting them

```python
plt.hist(np.random.randn(1000000,  2)[:, 1:2],  bins  =  np.linspace(-4, 4, 50)) 
plt.show()
```

```python
#plot of the rand_bi_gauss generated distribution
rand_bi_gauss_X, rand_bi_gauss_y = \
    rand_bi_gauss(n1 = 100, n2 = 200, mu1 = [1, 0], mu2 = [0, 0], sigmas1 = [0.2, 0.2], sigmas2 = [0.1, 0.1])
plot_2d(rand_bi_gauss_X,  rand_bi_gauss_y)
```

The previous function gives two gaussian distributions with given number of samples for each distribution and given params as mu and sigma

```python
#plot of the rand_tri_gauss generated distribution

rand_tri_gauss_X, rand_tri_gauss_y = rand_tri_gauss()
plot_2d(rand_tri_gauss_X, rand_tri_gauss_y)
```

The previous function gives three gaussian distributions with given number of samples for each distribution and given params as mu and sigma

```python
#plot of the rand_checkers generated distribution
rand_checkers_X, rand_checkers_y = rand_checkers()
plot_2d(rand_checkers_X,  rand_checkers_y)
```
