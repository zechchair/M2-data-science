```python
from sklearn.metrics import silhouette_score,calinski_harabasz_score

scores_K_1 = []
scores_K_2 = []
for i in range(2,6):
    model = GaussianMixture(n_components=i, random_state=0,covariance_type="diag").fit(x1)
    scores_K_1 += [silhouette_score(x1,model.predict(x1))]
    scores_K_2 += [calinski_harabasz_score(x1,model.predict(x1))]

axis_k = np.arange(2, 6, 1)
fig, axs = plt.subplots(1,2,figsize=(20, 10))
fig.suptitle('evaluation of temp data by  Silhouette_score with Calinski')
axs[0].plot(axis_k,scores_K_1)
axs[1].plot(axis_k,scores_K_2)
axs[0].set_title('Silhouette_score')
axs[1].set_title('Calinski')
```

```python


scores_K_1 = []
scores_K_2 = []
scores_K_3 = []

for i in range(2,6):
    model = GaussianMixture(n_components=i, random_state=0,covariance_type="diag").fit(x2)
    scores_K_1 += [silhouette_score(x2,model.predict(x2))]
    scores_K_2 += [calinski_harabasz_score(x2,model.predict(x2))]
    scores_K_3 += [adjusted_rand_score(y2.flatten(),model.predict(x2))]

axis_k = np.arange(2, 6, 1)
fig, axs = plt.subplots(1,3,figsize=(20, 10))
fig.suptitle('evaluation of mnist data by  Silhouette_score with Calinski then adjusted_rand')
axs[0].plot(axis_k,scores_K_1)
axs[1].plot(axis_k,scores_K_2)
axs[2].plot(axis_k,scores_K_3)
axs[0].set_title('Silhouette_score')
axs[1].set_title('Calinski')
axs[2].set_title('adjusted_rand')
```
